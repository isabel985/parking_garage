# Parking Garage Project

Your assignment for today is to create a parking garage class to get more familiar with Object Oriented Programming(OOP). 

Your parking gargage class should have the following methods:
- takeTicket
   - This should decrease the amount of tickets available by 1
   - This should decrease the amount of parkingSpaces available by 1
- payForParking
   - Display an input that waits for an amount from the user and store it in a variable
   - If the payment variable is not empty then ->  display a message to the user that their ticket has been paid and they have 15mins to leave
   - This should update the "currentTicket" dictionary key "paid" to True
-leaveGarage
   - If the ticket has been paid, display a message of "Thank You, have a nice day"
   - If the ticket has not been paid, display an input prompt for payment
      - Once paid, display message "Thank you, have a nice day!"
   - Update parkingSpaces list to increase by 1
   - Update tickets list to increase by 1

You will need a few attributes as well:
- tickets -> list
- parkingSpaces -> list
- currentTicket -> dictionary

```python
import uuid

class User:
    def __init__(self):
        self.ticket_id = uuid.uuid4()
        self.current_ticket = {"paid": False}

class Parking_Garage:
    def __init__(self):
        self.tickets = 100
        self.parking_spaces = 100
        self.users_in_garage = []
        
    def take_ticket(self):
        current_user = User()
        
        print(f"Your ticket id is: {current_user.ticket_id}")
        self.users_in_garage.append(current_user)
        print("===================================================================")
        
        print(f"Cars in garage: {len(self.users_in_garage)}")
        
        self.tickets -= 1
        self.parking_spaces -= 1
        
        print(f"Tickets left in machine: {self.tickets}")
        print(f"Parking Spaces left in parking garage: {self.parking_spaces}")
    
    def pay_for_parking(self):
        ask = input("Please give me your ticket id number: ")
        
        if not self.users_in_garage:
            print("Not a valid response.")
        
        elif self.users_in_garage:
             for user in self.users_in_garage:
                if ask == str(user.ticket_id):
                    def leave_garage(self):
                        self.users_in_garage.remove(user)
                        print(f"Cars in garage: {len(self.users_in_garage)}")

                        self.tickets += 1
                        print(f"Tickets left in parking garage: {self.tickets}")

                        self.parking_spaces += 1
                        print(f"Parking Spaces left in parking garage: {self.parking_spaces}")

                        print("Thank you, have a nice day! ")

                    while not user.current_ticket["paid"]:
                        hours = float(input("How many hours did you stay? Please round to the nearest hour. "))
                        print("You owe ${:.2f} ".format(round(hours)))
                        
                        while True:
                            payment = float(input("Plase enter your payment amount. "))
                            
                            if payment < hours:
                                print("Payment declined. Please enter the full amount. ")

                            elif payment >= hours:
                                print("===================================================================")
                                print("Your ticket has been paid. You have 15 minutes to leave. ")
                                leave_garage(self)
                                break
                        break

class Machine:
    def run(self):
        parking_garage = Parking_Garage()
        user = User()
        
        while True:
            ask = input("If entering the garage, type 'E'. If leaving the garage, type 'L'. ")
            print("===================================================================")
            if ask.upper() == "E":
                parking_garage.take_ticket()
            elif ask.upper() == "L":
                parking_garage.pay_for_parking()
            else:
                print("Please type 'E' or 'L'. ")
            
machine = Machine()
machine.run()
```
